# rand.function

```python
import numpy as np
```

```python
np.random.rand(3, 2)
```

###### This shows me the shape of the array
###### 3 means - how many rows is in my array and 2  - how many columns array has.
###### When there's no numbers in the bracket () "np.random.rand()" it gives me just a number. Rand is just generating random number.


```python
np.random.rand(2, 4, 3)
```

###### Here I have two-dimensional arrays, 4 rows in each array and 3 columns in each array.
